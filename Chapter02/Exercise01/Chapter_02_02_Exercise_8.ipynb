```python
import numpy as np #numerical computation
import pandas as pd #data wrangling
import matplotlib.pyplot as plt #plotting package
#Next line helps with rendering plots
%matplotlib inline
import matplotlib as mpl #add'l plotting functionality
mpl.rcParams['figure.dpi'] = 400 #high res figures
```

# Generating Synthetic Data

```python
from numpy.random import default_rng
rg = default_rng(12345)
```

```python
X = rg.uniform(low=0.0, high=10.0, size=(1000,))
X[0:10]
```

```python
type(X)
```

# Data for a Linear Regression

```python
slope = 0.25
intercept = -1.25
y = slope * X + rg.normal(loc=0.0, scale=1.0, size=(1000,)) + intercept
```

```python
mpl.rcParams['figure.dpi'] = 400
```

```python
plt.scatter(X,y,s=1)
plt.xlabel('X')
plt.ylabel('y')
```

# Exercise 8: Linear regression in Scikit-Learn

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg = LinearRegression(fit_intercept=True, normalize=False,
                           copy_X=True, n_jobs=None)
```

```python
lin_reg
```

```python
lin_reg.fit(X.reshape(-1,1), y)
```

```python
print(lin_reg.intercept_)
print(lin_reg.coef_)
```

```python
y_pred = lin_reg.predict(X.reshape(-1,1))
```

```python
plt.scatter(X,y,s=1)
plt.plot(X,y_pred,'r')
plt.xlabel('X')
plt.ylabel('y')
```
